### 模型效果对比

```python
def val_model(clf, X_train, x_test, y_train, y_test):
    clf.fit(X_train, y_train) # 拟合数据
    y_pred = clf.predict(x_test) # 预测数据
    print('classification_report\n', metrics.classification_report(y_test, y_pred, digits=4)) # 打印各类指标
    cf_mat = metrics.confusion_matrix(y_test, y_pred) # 获取混淆矩阵
    # 对混淆矩阵进行可视化
    hunxiao = pd.DataFrame(cf_mat)
    plt.subplots(figsize = (5, 5))
    sns.heatmap(hunxiao,annot=True,fmt = '.20g')
    plt.xlabel('predict') 
    plt.ylabel('true')
    plt.title('confusion matrix')
    plt.show()
    return clf

lr_clf = val_model(LogisticRegression(), X_train, X_test, y_train, y_test) # 对逻辑回归模型进行训练和测试
dt_clf = val_model(DecisionTreeClassifier(max_depth=2), X_train, X_test, y_train, y_test) # 对决策树模型进行训练和测试
rf_clf = val_model(RandomForestClassifier(max_depth=2, n_estimators=20), X_train, X_test, y_train, y_test) # 对随机森林模型进行训练和测试
ada_clf = val_model(AdaBoostClassifier(learning_rate=0.1,n_estimators = 20), X_train, X_test, y_train, y_test) # 对Adaboost模型进行训练和测试
knn_clf = val_model(KNeighborsClassifier(n_jobs=20,n_neighbors=8), X_train, X_test, y_train, y_test) # 对KNN模型进行训练和测试
```

```python
# 取单次预测的结果，绘制ROC曲线
def draw_roc(y_label, y_pre, name):
    fpr, tpr, thersholds = metrics.roc_curve(y_label, y_pre, pos_label=1)
    roc_auc = metrics.auc(fpr, tpr)
    plt.plot(fpr, tpr, label='{} ROC (area = {:.3f})'.format(name, roc_auc), lw=2)

def skl_model_choose(X_train, X_test, Y_train, Y_test, clf, name):
    X_train, X_test = np.array(X_train, dtype = int), np.array(X_test, dtype = int)
    Y_train, Y_test = np.array(Y_train, dtype = int), np.array(Y_test, dtype = int)
    print(type(clf))
    clf.fit(X_train, Y_train)
    y_pred = clf.predict(X_test)
    y_pred_pro = clf.predict_proba(X_test)[:, 1]
    acc = metrics.accuracy_score(Y_test, y_pred)
    print('Accuracy: ',acc)
    cf_mat = metrics.confusion_matrix(Y_test, y_pred)
    print('Confusion matrix: \n', cf_mat)
    clf_auc = metrics.roc_auc_score(Y_test,y_pred_pro)
    draw_roc(Y_test, y_pred_pro, name)
    print('auc : {}'.format(clf_auc))
    return acc, clf_auc, cf_mat

#lr_acc, lr_clf_auc, lr_cf_mat = skl_model_choose(X_train, X_test, y_train, y_test, LogisticRegression(), "lr")
dt_acc, dt_clf_auc, dt_cf_mat = skl_model_choose(X_train, X_test, y_train, y_test, DecisionTreeClassifier(max_depth=4), "dt")
rf_acc, rf_clf_auc, rf_cf_mat = skl_model_choose(X_train, X_test, y_train, y_test, RandomForestClassifier(max_depth=8, n_estimators=170, random_state=2022), "rf")
ada_acc, ada_clf_auc, ada_cf_mat = skl_model_choose(X_train, X_test, y_train, y_test, AdaBoostClassifier(learning_rate=0.1,n_estimators = 170), "ada")
#svc_acc, svc_clf_auc, svc_cf_mat = skl_model_choose(x_train, x_test, y_train, y_test, SVC(tol = 0.0001, kernel='linear'), "svc")
knn_acc, knn_clf_auc, knn_cf_mat = skl_model_choose(X_train, X_test, y_train, y_test, KNeighborsClassifier(n_jobs=20,n_neighbors=2), "knn")


plt.xlim([-0.05, 1.05])  
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.ensemble import VotingClassifier

estimators = []
estimators.append(('DecisionTreeClassifier', DecisionTreeClassifier(max_depth=10)))
estimators.append(('RandomForestClassifier', RandomForestClassifier(max_depth=8, n_estimators=170, random_state=2022)))
estimators.append(('AdaBoostClassifier', AdaBoostClassifier(learning_rate=0.1,n_estimators = 170)))

voting = VotingClassifier(estimators = estimators, voting='soft')
vote_acc, vote_clf_auc, vote_cf_mat = skl_model_choose(X_train,X_test, y_train, y_test, voting, "vote")
plt.xlim([-0.05, 1.05])  
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')  
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.show()
```
